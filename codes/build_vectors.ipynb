## **Import packages**

```python
import pandas as pd
from dask import dataframe as dd
import multiprocessing as mp
from sentence_transformers import SentenceTransformer
import os
import warnings

warnings.filterwarnings('ignore')
```

## **Get data**

```python
# Mount the files in drive folder
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/TIS_Q&A.csv')
print(df.shape)
df.head(3)
```

## **Embed Questions**

```python
# Load Bert sentence transformer
model_minilm = SentenceTransformer('all-MiniLM-L6-v2')

# Embed questions
embeddings_minilm = model_minilm.encode(list(df.question))
df['embedded_questions'] = embeddings_minilm.tolist()

df.head(5)
```

## **Export to .csv**

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/TIS_Embedded_Q&A.csv', index=False)
```
